Ques 6. Explain the concept of generators in Python and how they are defined.

Ans. A generator is a function that returns an iterable set of values. Generators avoid creating a list or other data structure in memory to store the values as they are produced, instead replacing them one by one over time. This makes them more efficient than loops for large datasets and allows for code that runs in linear time.

Generators are built using the yield keyword, which is used to indicate that a value should be returned from the function.

Defining a generator

- Define a function with the def keyword.
- Use the yield keyword within the function body to produce a sequence of values.
- The function becomes a generator when it contains yield.


```python
def even_numbers(n):
  for i in range(n):
    if i % 2 == 0:
      yield i

# Get the generator object
even_generator = even_numbers(10)

# Iterate using a for loop
for num in even_generator:
  print(num)
```

Ques 9. Explain the purpose and usage of the `map()` function in Python.

Ans. The map() function is used to execute a given function to each element of an iterable (such as lists, tuples, etc.).The map() function takes two arguments:

Syntax:
map(function, iterables)

function - a function to which map passes each element of given iterable.
iterables - iterables which is to be mapped such as lists, tuples, etc.

```python
# program to returns the square of a numbers in a list
numbers = [2,4,6,8]
def square(number):
  return number * number

squared_numbers = map(square, numbers) # applying square() to each item of the numbers list
result = list(squared_numbers) #list for printing
print(result)
```

Ques 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python.

Ans. These are the built-in functions of Python and used in different ways to manipulate iterables like list, tuple etc in a different purpose, such as

map() function is used to apply a given function to each element of an iterable and returns a new iterable with the results.
filter() function is used to filter elements from an iterable based on a given condition or function and returns a new iterable with the filtered elements.
reduce() function is used to apply a given function to the elements of an iterable in a cumulative way, reducing the iterable to a single value.

In general, we can say that map() applies a given function to each item of an array and returns a new array where filter() creates a new array with all elements that pass a test implemented by the provided function. And reduce() applies a function against an accumulator and each element in the array (from left to right) to reduce it to a single output value.

```python
# This prompt is asking to import the `reduce` function from the `functools` module
from functools import reduce
number = [47, 11, 42, 13]
sum = reduce(lambda x, y: x+y, number)
print ("sum of elements in ", number, "is :", sum)
```

In this example, we first imported the reduce() method from the functools module. We then defined a list of numbers and used the reduce() method to find their sum. We passed a lambda function as the first argument to the reduce() method. This lambda function takes two arguments x and y and returns their sum. The reduce() method applies this lambda function cumulatively to all elements in the list and returns their sum.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_even_nums(numbers):
    even = 0
    for num in numbers:
        if num % 2 == 0 :
            even = even + num
    return even

list1 = input ("Enter the list of numbers with space or comma:")
numbers = [int (num) for num in list1.replace(',' , '').split()]
result = sum_of_even_nums(numbers)
print("Sum of even numbers in the list is: ", result)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibonacci_series(n):
    fib_series_list = [0, 1]
    for i in range(2, n):
        fib_series_list.append(fib_series_list[-1] + fib_series_list[-2])
    return fib_series_list

# Example usage:
n = int(input(print("Enter a positive number")))
result = fibonacci_series(n)
print("Fibonacci series of ", n , "is :", result)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_generator(base, exponent):
    result = 1
    for i in range(exponent + 1):
        yield result
        result *= base

# Accept input from the user
base = int(input("Enter the base number: "))
exponent = int(input("Enter the exponent: "))

# Create the generator object
power_gen = power_generator(base, exponent)

# Generate and print the powers
print(f"Powers of {base} up to exponent {exponent}:")
for power in power_gen:
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(file_path):
  """Reads a file line by line and yields each line as a string.

  Args:
    file_path: The path to the file to read.

  Yields:
    The next line in the file as a string.
  """

  with open(file_path, 'r') as file:
    for line in file:
      yield line.rstrip('\n')  # Remove trailing newline character

# Example usage:
file_path = 'your_file.txt'
for line in read_file_lines(file_path):
  print(line)
```
